```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 20)
x.shape
```

```python
xx, yy = np.meshgrid(x, x)
xx.shape, yy.shape
```

```python
sns.heatmap(xx)
```

```python
X = xx.reshape(xx.shape[0]*xx.shape[1], 1)
X.shape
```

```python
X = np.hstack((X, yy.reshape(yy.shape[0]*yy.shape[1], 1)))
X.shape
```

```python
w = np.array([5, 3])
w_0 = -10

y_hat = X @ w + w_0
y_hat.shape
```

```python
y_hat_reshape = y_hat.reshape(xx.shape)
y_hat_reshape
```

```python
plt.contourf(xx, yy, y_hat_reshape)
```

```python
plt.contourf(xx, yy, y_hat_reshape>0)
```
